```python
import json
def add_node(root, path, text, filePaths):
    current = root
    elements = path.strip('/').split('/')
    for element in elements:
        if '[' in element:
            node_type, instance = element.split('[')
            instance = int(instance.strip(']'))
        else:
            node_type = element
            instance = 0
        
        found = False
        for child in current['children']:
            if child['type'] == node_type and child['instance'] == instance:
                current = child
                found = True
                break
        if not found:
            new_node = {
                'type': node_type,
                'instance': instance,
                'Text' : text,
                'children': [],
                'filePaths': '',
                'Path':path
            }
            current['children'].append(new_node)
            current = new_node

with open('/workspace/yule/PDFTriage/haigong/structuredData.json') as file:
    root_node = {'type': 'root', 'instance': 0, 'children': []}

    data = json.load(file)
    elements = data['elements']
    for ele in elements[1000:1100]:
        path = ele.get('Path', '')
        text = ele.get('Text', '')
        filePaths = ele.get('filePaths', '')
        try:
            add_node(root_node, path, text, filePaths)
        except:
            print(path)
            

# Printing the constructed tree
print(json.dumps(root_node, indent=4))
```
