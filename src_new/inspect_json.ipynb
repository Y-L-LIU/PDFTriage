```python
from dataclasses import dataclass, field
from typing import Self

from dataclasses_json import dataclass_json


@dataclass_json
@dataclass
class Node:
    path: str
    text: str
    depth: int
    children: list[Self] = field(default_factory=list)
```

```python
doc: Node = Node('//Document', '', 0)
stack: list[Node] = [doc]

for elem in elements:
    path = elem['Path']
    text = elem['Text'] if 'Text' in elem else ''
    ps = path.split('/')
    basename = ps[-1]
    if basename[0] == 'H':
        # is heading
        depth = int(basename[1])
        while depth <= stack[-1].depth:
            stack.pop()
    else: # not a heading
        depth = stack[-1].depth + 1
    node = Node(path, text, depth)
    stack[-1].children.append(node)
    if basename[0] == 'H':
        stack.append(node)
```
